This notebook is to test out the performance of leverage ETF compared with regular ETF

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

%matplotlib inline
```

```python
bull_market = 0.001
volatility = 0.017
days = 300

# np.random.seed(179)
daily_change = np.random.normal(loc=bull_market, scale=volatility, size=days)
print(daily_change[:10])
sns.distplot(daily_change, norm_hist=False)
```

```python

ETF_chart = np.ones(days)
ETF_3X_chart = np.ones(days)
for value in np.arange(ETF_chart.size):
    if value == 0:
        continue
    else:
        ETF_chart[value] = ETF_chart[value-1] * (1 + daily_change[value-1])
for value in np.arange(ETF_chart.size):
    if value == 0:
        continue
    else:
        ETF_3X_chart[value] = (ETF_3X_chart[value-1] * (1 + daily_change[value-1]*3))
    
date = range(days)
print(ETF_chart.size)
print(ETF_3X_chart.size)
# plt.subplot()`

fig, (ax_comb, ax_ETF, ax_ETF3X) = plt.subplots(nrows=3, sharex=True, num='Volatility', figsize = (13,10))
ax_comb.plot(date,ETF_chart)
ax_comb.plot(date,ETF_3X_chart, color = 'r')
ax_ETF.plot(date,ETF_chart)
ax_ETF3X.plot(date,ETF_3X_chart, color = 'r')
```
